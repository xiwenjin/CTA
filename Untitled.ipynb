# Indicators

```python
class indicators:
    def __init__(self):
        pass
        
    def EMA(self, price, days):       
        if not isinstance(price, pd.core.series.Series):
            raise Exception('The input prices should be object of pandas with datetime as index.')
            
        if price.index[0] < price.index[-1]:
            price = price.sort_index()      
        
        return price.ewm(span=days).mean()
    
    def MACD(self, price, fast, slow, sig):
        macd = self.EMA(price, fast) - self.EMA(price, slow)
        avg = self.EMA(macd, sig)
        hist = macd - avg
        
        return macd, avg, hist
    
    def signal_reverse(self, price, ds):
        pos, buy, sell = False, [], []
        
        ds = ds.diff(1)
        for i in ds.index[1:]:
            if not pos:
                if ds[i] > 0:
                    buy.append([i, price[i]])
                    sell.append([i, None])
                    pos = '+'
                elif ds[i] < 0:
                    sell.append([i, price[i]])
                    buy.append([i, None])
                    pos = '-'
                else:
                    continue
            else:
                if pos == '+':
                    buy.append([i, None])
                    if ds[i] >= 0:
                        sell.append([i, None])
                    if ds[i] < 0:
                        sell.append([i, price[i]])
                        pos = '-'
                else:
                    sell.append([i, None])
                    if ds[i] <= 0:
                        buy.append([i, None])
                    if ds[i] > 0:
                        buy.append([i, price[i]])
                        pos = '+'
                        
        b_sig, s_sig = pd.DataFrame({'date': ds.index[0], 'Close': [None]}), \
                       pd.DataFrame({'date': ds.index[0], 'Close': [None]})          
        
        buy = pd.concat([b_sig, pd.DataFrame({'date': list(zip(*buy))[0], 
                                            'Close': list(zip(*buy))[1]})]).set_index('date')
        sell = pd.concat([s_sig, pd.DataFrame({'date': list(zip(*sell))[0],
                                              'Close': list(zip(*sell))[1]})]).set_index('date')
        return buy, sell
    
```

# simple back test

```python
class myDataFeed(bt.feeds.PandasData):
    lines = ('macd', 'signal', 'hist')
    params = (('macd', -1), ('signal', -1), ('hist', -1), 
              ('macd_ftd', -1), ('signal_ftd', -1), ('hist_ftd', -1),)


class myStrat(bt.Strategy):
    params = (('fast', 12), ('slow', 26), ('sig', 9), ('col', 'hist'))
    
    def log(self, msg, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        print(str(dt.isoformat()) + ', ' + msg)
    
    def __init__(self):
        '''
        loc: which column to use as indicator
        '''
        self.dataclose = self.datas[0].close
        self.ind = self.datas[0]._getline(self.p.col)
        self.order = None
        
        self.buy_price = None
        self.buy_comm = None
        self.sell_price = None
        self.sell_comm = None
        
        self.const_size = 1
        
        # count the iteration of next()
        self.ct = 0
        
        
    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return
        
        if order.status in [order.Completed]:
            if order.isbuy():
                # print order info
                self.log('Buy exe @ ' + str(np.round(order.executed.price, 2)) +\
                         ', Cost: ' + str(np.round(order.executed.value, 2)) +\
                         ', Commision: ' + str(np.round(order.executed.comm, 2)) +\
                         ', Position: ' + str(self.position.size))
                
                # record executed price
                self.buy_price = order.executed.price
                self.buy_comm = order.executed.comm
            else:
                self.log('Sell exe @ ' + str(np.round(order.executed.price, 2)) +\
                         ', Cost: ' + str(np.round(order.executed.value, 2)) +\
                         ', Commision: ' + str(np.round(order.executed.comm, 2)) +\
                         ', Position: ' + str(self.position.size))
                
                # record executed price
                self.sell_price = order.executed.price
                self.sell_comm = order.executed.comm
                
            self.bar_executed = len(self)
                
        if order.status in [order.Canceled, order.Margin, order.Rejected]:
            self.log('Order Canceled/Margin/Rejected')
            
        self.order = None
            
    
    def notify_trade(self, trade):
        if not trade.isclosed:
            return
        
        self.log('Gross PnL: ' + str(np.round(trade.pnl, 2)) +\
                 ', Net PnL: ' + str(np.round(trade.pnlcomm, 2)))
        
    
    def next(self):
        
        self.ct += 1
        self.log('Close price: ' + str(self.dataclose[0]) +\
                 ', Balance: ' + str(np.round(self.stats.broker.value[0], 2)) +\
                 ', Indicator: ' + str(np.round(self.ind[0], 2)))

        # if order is submitted or pending, we do nothing.
        if self.order:
            return
        
        if not self.position:
            if self.ct > 1: # when self.ct = 0, self.ind[-1] is last indicator in the data feed
                if self.ind[0] > self.ind[-1]:
                    self.log('Signal: Open Buy @ ' + str(self.dataclose[0]))
                    self.order = self.buy(size=self.const_size)
                elif self.ind[0] < self.ind[-1]:
                    self.log('Signal: Open Sell @ ' + str(self.dataclose[0]))
                    self.order = self.sell(size=self.const_size)
                else: 
                    return
            else:
                return 
        else:        
            if self.ind[0] > self.ind[-1]:
                if self.position.size < 0:
                    self.log('Signal: Exit & Buy @ ' + str(self.dataclose[0]))
                    self.order = self.buy(size=self.const_size * 2)
                else:
                    return
                    
            elif self.ind[0] < self.ind[-1]:
                if self.position.size > 0:                    
                    self.log('Signal: Exit & Sell @ ' + str(self.dataclose[0]))
                    self.order = self.sell(size=self.const_size * 2)
                else:
                    return
            else:
                return
        
        

```

```python
yr = 2021
start, end = pd.datetime(yr-1, 1, 1), pd.datetime(yr, 12, 31)

d50, d300, d500 = sse50[start: end], csi300[start: end], csi500[start: end]
dwti, dbrt = wti[start: end], brt[start: end]

idx = d500[:]

ind = indicators()
fast = 12
slow = 26
sig = 9

macd, avg, hist = ind.MACD(idx.Close, fast, slow, sig)
idx['macd'], idx['signal'], idx['hist'] = macd, avg, hist

buy, sell = ind.signal_reverse(idx.Close, hist)
```

```python
cbro = bt.Cerebro(stdstats=False)
cbro.addstrategy(myStrat, col='hist')

# data = bt.feeds.PandasData(dataname=idx,
#                            fromdate=start,
#                            todate=end,
#                            plot=False)

# init data feed with my own indicators
data = myDataFeed(dataname=idx,
                  fromdate=start,
                  todate=end,
                  plot=False)
# data.plotinfo.subplot = True


cbro.adddata(data)
cbro.addobserver(bt.observers.Broker)
cbro.addobserver(bt.observers.Trades)
cbro.addobserver(bt.observers.BuySell)
cbro.addobserver(bt.observers.DrawDown)
cbro.addobserver(bt.observers.TimeReturn)

cbro.broker.setcommission(commission=10, margin=True, mult=200)
# cbro.broker.set_slippage_perc(5e-3)

# assume order executed on the close of the bar that the same bar the signal is generated,
# as the indicator is calculated using open prices.
# cbro.broker.set_coc(True)
cbro.broker.setcash(1e+6)

print('Starting balance:', str(cbro.broker.getvalue()))

cbro.run()

print('Ending balance:', str(cbro.broker.getvalue()))
```

# with filtering

```python
# set up design matrix
# assume state process: x_t = x_{t-1} + w_t, w ~ N(0, Q)
# measurement process: y_t = x_t + e_t, e ~ N(0, R)
# inital state: x0 = first observation and p0 = std of close price during previous 1 month.
A = 1
B = 0
H = 1
Q = 1 #csi500.loc[pd.datetime(yr-1, 12, 1): pd.datetime(yr-1, 12, 31), 'Close'].std()
R = 100

# calc filtered macd and signal lines
idx = d500[:]
macd, avg, hist = ind.MACD(idx.Close, fast, slow, sig)
idx['macd'], idx['signal'], idx['hist'] = macd, avg, hist

x, P = idx.macd[0], 1
k_filter = kf.KalmanFilter(A, B, H, Q, R)
macd_ftd, _, _, _ = k_filter.filt(idx.macd, x, P)
macd_ftd = k_filter.to_array(macd_ftd)

x, P = idx.signal[0], 1
signal_ftd, _, _, _ = k_filter.filt(idx.signal, x, P)
signal_ftd = k_filter.to_array(signal_ftd)

hist_ftd = pd.Series(macd_ftd - signal_ftd, index=idx.index)
buy, sell = ind.signal_reverse(idx.Close, hist_ftd)

# change data feed with filtered values
idx['macd'], idx['signal'], idx['hist'] = macd_ftd, signal_ftd, hist_ftd
```
