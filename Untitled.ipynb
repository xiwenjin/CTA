# Indicators

```python
class indicators:
    def __init__(self):
        pass
        
    def EMA(self, price, days):       
        if not isinstance(price, pd.core.series.Series):
            raise Exception('The input prices should be object of pandas with datetime as index.')
            
        if price.index[0] < price.index[-1]:
            price = price.sort_index()      
        
        return price.ewm(span=days).mean()
    
    def MACD(self, price, fast, slow, sig):
        macd = self.EMA(price, fast) - self.EMA(price, slow)
        avg = self.EMA(macd, sig)
        hist = macd - avg
        
        return macd, avg, hist
    
    def signal_reverse(self, price, ds):
        pos, buy, sell = False, [], []
        
        ds = ds.diff(1)
        for i in ds.index[1:]:
            if not pos:
                if ds[i] > 0:
                    buy.append([i, price[i]])
                    sell.append([i, None])
                    pos = '+'
                elif ds[i] < 0:
                    sell.append([i, price[i]])
                    buy.append([i, None])
                    pos = '-'
                else:
                    continue
            else:
                if pos == '+':
                    buy.append([i, None])
                    if ds[i] >= 0:
                        sell.append([i, None])
                    if ds[i] < 0:
                        sell.append([i, price[i]])
                        pos = '-'
                else:
                    sell.append([i, None])
                    if ds[i] <= 0:
                        buy.append([i, None])
                    if ds[i] > 0:
                        buy.append([i, price[i]])
                        pos = '+'
                        
        b_sig, s_sig = pd.DataFrame({'date': ds.index[0], 'Close': [None]}), \
                       pd.DataFrame({'date': ds.index[0], 'Close': [None]})          
        
        buy = pd.concat([b_sig, pd.DataFrame({'date': list(zip(*buy))[0], 
                                            'Close': list(zip(*buy))[1]})]).set_index('date')
        sell = pd.concat([s_sig, pd.DataFrame({'date': list(zip(*sell))[0],
                                              'Close': list(zip(*sell))[1]})]).set_index('date')
        return buy, sell
    
```

```python
class myDataFeed(bt.feeds.PandasData):
    lines = ('macd', 'signal', 'hist')
    params = (('macd', -1), ('signal', -1), ('hist', -1), 
              ('macd_ftd', -1), ('signal_ftd', -1), ('hist_ftd', -1),)
    
class myTrades(bt.observers.Trades):
    plotlines = dict(
        pnlplus = dict(_name='positive', marker='+', markersize=8, 
                       color='green', fillstyle='full'),
        pnlminus = dict(_name='negative', marker='*', markersize=8, 
                        color='red', fillstyle='full'),
    )
    
class myBuySell(bt.observers.BuySell):
    params = (('bardist', 1.5), )
    plotlines = dict(
        buy = dict(marker='^', markersize=5, color='green'),
        sell = dict(marker='v', markersize=5, color='red'),
        stop = dict(marker='x', markersize=5, color='black'))


class myStrat(bt.Strategy):
    params = (('fast', 12), ('slow', 26), ('sig', 9), ('col', 'hist'), ('stop_loss_pct', .05))
    
    def log(self, msg, dt=None):
        dt = dt or self.datas[0].datetime.date(0)
        print(str(dt.isoformat()) + ', ' + msg)
    
    def __init__(self):
        '''
        '''
        self.dataclose = self.datas[0].close
        self.ind = self.datas[0]._getline(self.p.col)
        
        self.order = None
        self.stop_loss = None
        
        self.exe_price = None
        self.exe_comm = None
        self.last_exe_price = {'last': None, 'crt': None}

        self.const_size = 1
        self.flip_size = self.const_size * 2
        self.td = {'last': 'flat', 'crt': 'flat'}
        
        # count the iteration of next()
        self.ct = 0
        
        
    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return
        
        if order.status in [order.Completed]:
            self.last_exe_price['last'] = self.last_exe_price['crt']
            self.last_exe_price['crt'] = str(np.round(order.executed.price, 2))
            
            if order.isbuy():
                # print order info
                self.log(str(self.ct) + ', ' + order.info['name'] +\
                         ' exe @ ' + str(np.round(order.executed.price, 2)) +\
                         ', Cost: ' + str(np.round(order.executed.value, 2)) +\
                         ', Commision: ' + str(np.round(order.executed.comm, 2)) +\
                         ', Position: ' + str(self.position.size))
                
                # record executed price
                self.exe_price = order.executed.price
                self.exe_comm = order.executed.comm
            else:
                self.log(str(self.ct) + ', ' + order.info['name'] +\
                         ' exe @ ' + str(np.round(order.executed.price, 2)) +\
                         ', Cost: ' + str(np.round(order.executed.value, 2)) +\
                         ', Commision: ' + str(np.round(order.executed.comm, 2)) +\
                         ', Position: ' + str(self.position.size))
                
                # record executed price
                self.exe_price = order.executed.price
                self.exe_comm = order.executed.comm
                
            self.bar_executed = len(self)
                
        if order.status in [order.Margin, order.Rejected]:
            self.log('Order Margin/Rejected')
        
        if order.status == order.Cancelled:
            self.log(str(self.ct) + ', ' + order.info['name'] +\
                     ' canceled on: ' + str(self.datas[0].datetime.date(0)))
            
        self.order = None
            
    
    def notify_trade(self, trade):
        if not trade.isclosed:
            return
        
        self.log(str(self.ct) + ', ' + 'Trade closed @ ' +\
                 str(np.round(self.exe_price, 2)) +\
                 ', Last exe: ' + self.last_exe_price['last'] +\
                 ', Gross PnL: ' + str(np.round(trade.pnl, 2)) +\
                 ', Net PnL: ' + str(np.round(trade.pnlcomm, 2)))
        
    
    def next(self):
        self.ct += 1
        self.log(str(self.ct) + ', close: ' + str(self.dataclose[0]) +\
                 ', open: ' + str(self.datas[0].open[0]) +\
                 ', Value: ' + str(np.round(self.stats.broker.value[0], 2)) +\
                 ', Cash: ' + str(np.round(self.stats.broker.cash[0], 2)) +\
                 ', Indicator: ' + str(np.round(self.ind[0], 2)))

        # if order is submitted or pending, we do nothing.
        if self.order:
            return
        
        # if it's the first data feed, do nothing.
        if self.ct == 1:
            return
        
        # to open position, for the initial and opening after stop loss/taking profit
        if self.position.size == 0:
            if self.ind[0] > self.ind[-1]:
                self.td['last'] = self.td['crt']
                self.td['crt'] = 'up'
                
                # open position after stop loss/taking profit, it can only go long
                # as crt is always 'up'
                if self.td['last'] != self.td['crt']:
                    self.order = self.buy(size=self.const_size, transmit=False)
                    self.order.addinfo(name='Open Buy-' + str(self.ct))
                    self.log(str(self.ct) + ', Signal: ' + self.order.info['name'] +\
                             ' @ ' + str(self.dataclose[0]))
                    # add stop loss
                    self.stop_loss = self.sell(size=self.const_size, exectype=bt.Order.Stop, 
                                               price=self.dataclose[0] * (1 - self.p.stop_loss_pct),
                                               parent=self.order)
                    self.stop_loss.addinfo(name='Stop loss sell-' + str(self.ct))
                else:
                    return
                
            elif self.ind[0] < self.ind[-1]:
                self.td['last'] = self.td['crt']
                self.td['crt'] = 'down'
                
                # open position after stop loss/taking profit, it can only go short
                # as crt is always 'down'
                if self.td['last'] != self.td['crt']:
                    self.order = self.sell(size=self.const_size, transmit=False)
                    self.order.addinfo(name='Open Sell-' + str(self.ct))
                    self.log(str(self.ct) + ', Signal: ' + self.order.info['name'] +\
                             ' @ ' + str(self.dataclose[0]))
                    # add stop loss
                    self.stop_loss = self.buy(size=self.const_size, exectype=bt.Order.Stop, 
                                              price=self.dataclose[0] * (1 + self.p.stop_loss_pct),
                                              parent=self.order)
                    self.stop_loss.addinfo(name='Stop loss buy-' + str(self.ct))
                else:
                    return
            # indicator no chg --> do nothing
            else:
                return
        # to flip position, when self.position == True
        else:
            if self.ind[0] > self.ind[-1]:
                self.td['last'] = self.td['crt']
                self.td['crt'] = 'up'
                
                if self.position.size < 0:
                    if self.stop_loss:
                        self.cancel(self.stop_loss)
                        
                    self.order = self.buy(size=self.flip_size, transmit=False)
                    self.order.addinfo(name='Flip Buy-' + str(self.ct))
                    self.log(str(self.ct) + ', Signal: ' + self.order.info['name'] +\
                             ' @ ' + str(self.dataclose[0]))
                    # add stop loss 
                    self.stop_loss = self.sell(size=self.const_size, exectype=bt.Order.Stop, 
                                               price=self.dataclose[0] * (1 - self.p.stop_loss_pct),
                                               parent=self.order)
                    self.stop_loss.addinfo(name='Stop loss sell-' + str(self.ct))
                else:
                    return
                    
            elif self.ind[0] < self.ind[-1]:
                self.td['last'] = self.td['crt']
                self.td['crt'] = 'down'
                                        
                if self.position.size > 0:
                    if self.stop_loss:
                        self.cancel(self.stop_loss)
                        
                    self.order = self.sell(size=self.flip_size, transmit=False)
                    self.order.addinfo(name='Flip Sell-' + str(self.ct))
                    self.log(str(self.ct) + ', Signal: ' + self.order.info['name'] +\
                             ' @ ' + str(self.dataclose[0]))
                    # add stop loss
                    self.stop_loss = self.buy(size=self.const_size, exectype=bt.Order.Stop, 
                                              price=self.dataclose[0] * (1 + self.p.stop_loss_pct),
                                              parent=self.order)
                    self.stop_loss.addinfo(name='Stop loss buy-' + str(self.ct))   
                else:
                    return
            else:
                return
        
        
```

```python
# set up design matrix
# assume state process: x_t = x_{t-1} + w_t, w ~ N(0, Q)
# measurement process: y_t = x_t + e_t, e ~ N(0, R)
# inital state: x0 = first observation and p0 = std of close price during previous 1 month.
A = 1
B = 0
H = 1
Q = 1 #csi500.loc[pd.datetime(yr-1, 12, 1): pd.datetime(yr-1, 12, 31), 'Close'].std()
R = 100

# calc filtered macd and signal lines
idx = idx_dict[prod][0]
macd, avg, hist = ind.MACD(idx.Close, fast, slow, sig)
idx['macd'], idx['signal'], idx['hist'] = macd, avg, hist

x, P = idx.macd[0], 1
k_filter = kf.KalmanFilter(A, B, H, Q, R)
macd_ftd, _, _, _ = k_filter.filt(idx.macd, x, P)
macd_ftd = k_filter.to_array(macd_ftd)

x, P = idx.signal[0], 1
signal_ftd, _, _, _ = k_filter.filt(idx.signal, x, P)
signal_ftd = k_filter.to_array(signal_ftd)

hist_ftd = pd.Series(macd_ftd - signal_ftd, index=idx.index)
buy, sell = ind.signal_reverse(idx.Close, hist_ftd)

# change data feed with filtered values
idx['macd'], idx['signal'], idx['hist'] = macd_ftd, signal_ftd, hist_ftd
```
